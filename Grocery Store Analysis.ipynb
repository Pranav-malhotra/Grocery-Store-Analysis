```python
df_inc = df.copy()
df_dec = df.copy()
```

```python
# What happens when we increase the price

def increase_price(df):
    df['Avg_Price_per_unit'] = df['Avg_Price_per_unit'] + (df['Avg_Price_per_unit'] * 1/100) # Incremement 1 percent on price
    df['Average_Profit_per_unit'] = df['Avg_Price_per_unit'] - df['Cost_per_unit']  # Profit Increases
    
    #df['Acquisition'] = df['Acquisition'] - (df['Incremental_acquisition'] / 10)
    df['Acquisition'] = df['Acquisition'] - (df['Incremental_acquisition'])  # Acquisition decreases
    df['Average_units_sold'] = df['Average_units_sold'] - (df['Increase_sale_volume'])  # Average Unit Sold Decreases
```

```python
# What happens when we decrease the price

def decrease_price(df):
    df['Avg_Price_per_unit'] = df['Avg_Price_per_unit'] - (df['Avg_Price_per_unit'] * 1/100) # Decremement 1 percent
    df['Average_Profit_per_unit'] = df['Avg_Price_per_unit'] - df['Cost_per_unit']  # Profit Decreases
    
    #df['Acquisition'] = df['Acquisition'] + (df['Incremental_acquisition'] / 10)
    df['Acquisition'] = df['Acquisition'] + (df['Incremental_acquisition'])  # Acquisition increases
    df['Average_units_sold'] = df['Average_units_sold'] + (df['Increase_sale_volume'])  # Average Unit sold Increases
```

```python
product_profit = pd.DataFrame() # DF with profit in 31 scenarios of all 250 products
```

```python
product_profit[0] = df['Average_Profit_per_unit'] * df['Average_units_sold'] # Keeping in account of default product profit
```

```python
# Increasing Price by 1% upto 20%
for i in range(20):
    increase_price(df_inc)
    profit_list = (df_inc['Average_Profit_per_unit'] * df_inc['Average_units_sold'])
    product_profit[i+2] = profit_list
```

```python
# Decreasing Price by 1% upto 10%

for i in range(10):
    decrease_price(df_dec)
    profit_list = (df_dec['Average_Profit_per_unit'] * df_dec['Average_units_sold'])
    product_profit[-1 - i] = profit_list
```

```python
product_profit
```

```python
product_profit_t = product_profit.T
```
